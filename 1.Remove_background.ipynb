```python
from rembg import remove
from PIL import Image

def process_image(input_path, output_path):
    """
    Removes the background from the input image and saves the result.

    Args:
        input_path (str): The file path to the input image.
        output_path (str): The file path to save the output image.
    """
    try:
        # Open the input image
        input_image = Image.open(input_path)

        # Remove the background
        output_image = remove(input_image)

        # Save the output image
        output_image.save(output_path)

        print(f"Image processed and saved to {output_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    # Define input and output paths
    image_input = "/content/dog2.png"
    image_output = "/content/output.png"

    # Process the image
    process_image(image_input, image_output)
```
